```python
import numpy as np
data=np.loadtxt('data_singlevar.txt',delimiter=',')
c1,c2=np.hsplit(data,2)
import sklearn
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(c1,c2)
from sklearn import linear_model
lr=linear_model.LinearRegression()
lr.fit(train_x,train_y)
test_y1=lr.predict(test_x)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(15, 15))
plt.scatter(train_x, train_y, color='red')
plt.plot(test_x, test_y1, color='black', linewidth=2)
plt.scatter(test_x, test_y, color='green')
plt.scatter(test_x, test_y1, color='blue')
```

```python
import sklearn.metrics as sm
print ("Mean absolute error = {}".format(round(sm.mean_absolute_error(test_y,test_y1), 2)))
print ("Mean squared error ={}".format( round(sm.mean_squared_error(test_y,test_y1), 2)))
print ("Median absolute error ={}".format( round(sm.median_absolute_error(test_y,test_y1), 2)))
print ("Explained variance score ={}".format(round(sm.explained_variance_score(test_y,test_y1), 2)))
print ("R2 score ={}".format( round(sm.r2_score(test_y,test_y1), 2)))
```
